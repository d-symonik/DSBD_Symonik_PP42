```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn import preprocessing
from sklearn.feature_selection import VarianceThreshold, mutual_info_regression, f_regression, SelectKBest, f_classif, \
    mutual_info_classif
```

```python
automobile = pd.read_csv(f'D:\pythonProject5\Automobile_data.csv',sep=',')
automobile.head(10)
```

Проведемо очистку та обробку  даних

```python
automobile.replace("?", np.nan, inplace=True)

numeric_cols = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
automobile[numeric_cols] = automobile[numeric_cols].apply(pd.to_numeric, errors='coerce')

automobile.dropna(axis=0, how='all', inplace=True)
automobile.head(10)
```

```python
categories = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', 'fuel-system']

fig, axes = plt.subplots(nrows=len(categories) // 2, ncols=2, figsize=(15, 20))

for i, feature in enumerate(categories):
    row = i // 2
    col = i % 2
    sns.countplot(x=feature, data=automobile, ax=axes[row, col])
    axes[row, col].tick_params(labelrotation=45)
    axes[row, col].set_title(f'Розподіл по {feature}')

plt.tight_layout()
plt.show()
```

```python
nums_params = automobile.select_dtypes(include=['float64', 'int64'])

X = nums_params.drop("price", axis=1)
y = nums_params["price"]

X.fillna(X.mean(), inplace=True)
y.fillna(y.mean(), inplace=True)

selector = VarianceThreshold()
variance = selector.fit(X)

fscore = SelectKBest(f_classif, k='all')
fscore.fit_transform(X, y)

lab = preprocessing.LabelEncoder()
mutual = SelectKBest(mutual_info_classif, k='all')
mutual.fit_transform(X, lab.fit_transform(y))

result_df = pd.DataFrame(index=X.columns)
result_df["Variance Threshold"] = variance.variances_
result_df['F-Score'] = fscore.scores_
result_df['MutualInformation'] = mutual.scores_

result_df.transpose()
```

```python
plt.figure(figsize=(18, 6))
sns.heatmap(result_df[["Variance Threshold"]].sort_values(by='Variance Threshold', ascending=False),annot=True, fmt=".3f", cmap='plasma')
plt.title('Variance Threshold')
plt.show()

plt.figure(figsize=(18, 6))
sns.heatmap(result_df[["F-Score"]].sort_values(by='F-Score', ascending=False), annot=True,
            fmt=".3f", cmap='plasma')
plt.title('F-Score')
plt.show()

plt.figure(figsize=(18, 6))
sns.heatmap(result_df[["MutualInformation"]].sort_values(by='MutualInformation', ascending=False), annot=True,
            fmt=".3f", cmap='plasma')
plt.title('MutualInformation')
plt.show()

```
